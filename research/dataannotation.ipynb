```python
import os
import pandas as pd

metadata_csv = "Metadata_ElectroCom61.csv"
label_folders = ["train/labels", "validation/labels", "test/labels"]

# Step 1: Get unique class IDs from YOLO labels
all_class_ids = set()
for folder in label_folders:
    if not os.path.exists(folder):
        continue
    for fname in os.listdir(folder):
        if fname.endswith(".txt"):
            path = os.path.join(folder, fname)
            with open(path, "r") as f:
                for line in f:
                    if line.strip():
                        class_id = int(line.split()[0])
                        all_class_ids.add(class_id)

all_class_ids = sorted(list(all_class_ids))
print(f"Unique class IDs found in labels: {all_class_ids}")


# Step 2: Extract all individual object names from CSV
df = pd.read_csv(metadata_csv)
df.columns = df.columns.str.strip()

unique_objects = set()
for obj_cell in df['OBJECT_NAME']:
    if pd.isna(obj_cell):
        continue
    for obj in obj_cell.split():  
        name = obj.split('_')[0].strip()
        unique_objects.add(name)

unique_objects = sorted(unique_objects)

id_to_name = {cid: name for cid, name in zip(all_class_ids, unique_objects)}

# Step 3: Final mapping
print("\nFinal YOLO class_id → component mapping:")
for cid in all_class_ids:
    label_name = id_to_name.get(cid, "UNKNOWN")
    print(f"{cid} → {label_name}")
```
